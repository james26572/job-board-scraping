```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import csv
from jobspy import scrape_jobs
from concurrent.futures import ThreadPoolExecutor, as_completed
```

```python
job_titles = ['software engineer','financial analyst','accountant','investment banker','product manager',
              'data scientist','project manager','sales manager','marketing specialist','web developer']
```

```python
results = pd.DataFrame({})

for job_title in job_titles:
    jobs = scrape_jobs(
    site_name=["linkedin"],
    search_term=job_title,
    location="London",
    results_wanted=100,
    hours_old=72, # (only Linkedin/Indeed is hour specific, others round up to days old)
    country_indeed='UK'  # only needed for indeed / glassdoor
)
    print(f"Found {len(jobs)} jobs")
    results = pd.concat([results,jobs],axis = 0)
    

```

```python
def get_page_content(url):
    import sys
    import ssl
    ssl._create_default_https_context = ssl._create_unverified_context
    
    if sys.version_info[0]==3:
        import urllib.request
        opener = urllib.request.build_opener(
            urllib.request.ProxyHandler(
                {'http': 'http://brd-customer-hl_a4d34049-zone-unblocker:4rc3q2njeh4a@brd.superproxy.io:22225',
                'https': 'http://brd-customer-hl_a4d34049-zone-unblocker:4rc3q2njeh4a@brd.superproxy.io:22225'}))
        return str(opener.open(url).read())
```

```python
def get_job_description(description_url):
    
    response = get_page_content(description_url)
    soup = BeautifulSoup(response,'html.parser')
    description = soup.find('div',class_ ='show-more-less-html__markup')
    
    cleaned_description = description.text.strip()
    
    return cleaned_description
```

```python
def fetch_descriptions(urls, max_workers=10):
    
    with ThreadPoolExecutor(max_workers=max_workers) as executor:
        future_to_url = {executor.submit(get_job_description, url): url for url in urls}
        results = []
        for future in as_completed(future_to_url):
            url = future_to_url[future]
            try:
                result = future.result()
                results.append(result)
            except Exception as e:
                print(f"Error fetching job description from {url}: {e}")
                results.append(None)
    return results
```

```python
results['description'] = fetch_descriptions(results.job_url)
```

```python
results.drop(columns=['interval', 'min_amount', 'max_amount',
       'currency', 'is_remote', 'emails','company_url',
       'company_url_direct', 'company_addresses', 'company_industry',
       'company_num_employees', 'company_revenue', 'company_description',
       'logo_photo_url', 'banner_photo_url', 'ceo_name', 'ceo_photo_url',], inplace = True)
```
